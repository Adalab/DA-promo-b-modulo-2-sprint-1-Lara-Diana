```python
# Sacad los valores únicos y su frecuencia de las variables numéricas. Almacenad los resultados en un diccionario. Uno para cada variable.
for i in nombres_columna:
    attacks_csv[i].value_counts()
   
    
```

```python
attacks_csv["Time"].value_counts()
```

```python
# Cambiad el nombre de las columnas para que todas estén en minúsculas y reemplazad los espacios por _.
columnas_nuevas = {col : col.lower() for col in attacks_csv.columns}
attacks_csv.rename(columns = columnas_nuevas, inplace=True)
columnas_nuevas2 = {col:col.replace(' ','-') for col in attacks_csv.columns}
attacks_csv.rename(columns = columnas_nuevas2, inplace=True)
attacks_csv
```

```python
# Eliminad las columnas de Unnamed: 22 y Unnamed: 23, Case Number.1, Case Number.2, original order, Investigator or Source, pdf, href formula.
attacks_csv.drop(['unnamed:-22','unnamed:-23','case-number.1','original-order','investigator-or-source','pdf','href-formula'],axis=1,inplace=True)
```

```python
attacks_csv
```

```python

# Establece la columna Case Number como índice del DataFrame.

attacks_csv.set_index(['case-number'],inplace=True)
attacks_csv
```

```python

# 3-Guarda el csv que no tiene esas columnas para seguir usándolo mañana.

attacks_csv.to_csv('./datos/attacks_modificado.csv')
```
